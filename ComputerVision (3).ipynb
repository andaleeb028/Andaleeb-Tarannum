## Step 1: Installing libraries

```python
from PIL import Image,ImageChops
import matplotlib.pyplot as plt
import numpy as np
from numpy import asarray
```

## Step 2: Reading colour image

```python
imgs = Image.open("fullnight.jpg")
plt.imshow(imgs)
```

```python
plt.show()
```

## Step 3: Converting Colour Image to Grayscale

```python
#Grayscale Image
Gimg = imgs.convert('L')
plt.imshow(Gimg,cmap='gray')
plt.show()
```

## Step 4: Converting image into numpy array

```python
# load the image and convert into
# numpy array
img = Image.open("fullnight.jpg")
numpydata = asarray(img)

# data
print(numpydata)
```

## Step 5: Determing the dimension of the array

```python
img= Image.open("fullnight.jpg")
np_img = np.array(img)
print(np_img.shape)
```

## Step 6: Determing the average of pixel values of the grayscale image

```python
average = np.average(np_img)
print(average)
```

## Step 7: Detecting day or night based on the threshold of 127

```python
threshold = 127
if average > threshold:
    print("Day")
else:
    print("Night")
```
