## Step 1: Importing libraries

```python
from PIL import Image,ImageChops
import matplotlib.pyplot as plt
import numpy as np
from numpy import asarray
```

## Step 2: Reading two colour images

```python
img1 = Image.open("fullnight.jpg")
plt.imshow(img1)
```

```python
img2 = Image.open("day.jpg")
plt.imshow(img2)
```

## Step 3: Converting Colour Image to Grayscale

```python
#Grayscale Image
Gimg = img1.convert('L')
plt.imshow(Gimg,cmap='gray')
plt.show()
```

```python
#Grayscale Image
Gimg = img2.convert('L')
plt.imshow(Gimg,cmap='gray')
plt.show()
```

## Step 4: Converting image into numpy array

```python
# convert into numpy array for img1
numpydata1 = asarray(img1)
# data
print(numpydata1)
```

```python
# convert into numpy array for img2
numpydata2 = asarray(img2)
# data
print(numpydata2)
```

## Step 5: Determing the dimension of the array

```python
#dimension of the image
np_img1 = np.array(img1)
print(np_img1.shape)
```

```python
#dimension of the image
np_img2 = np.array(img2)
print(np_img2.shape)
```

## Step 6: Determing the average of pixel values of the grayscale image

```python
average1 = np.average(np_img1)
print(average1)
```

```python
average2 = np.average(np_img2)
print(average2)
```

## Step 7: Detecting day or night based on the threshold of 100 for img1

```python
threshold = 100
if average1 > threshold:
    print("Day")
else:
    print("Night")
```

## Step 8: Detecting day or night based on the threshold of 100 for img2

```python
threshold = 100
if average2 > threshold:
    print("Day")
else:
    print("Night")
```
